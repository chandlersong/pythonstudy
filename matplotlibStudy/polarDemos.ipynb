```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute pie slices
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(111, projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0,alpha=0.5)

plt.show()
```

官网上面的雷达图。主要是理解了雷达图。大致怎么来画。
1. 告诉matplotlib。用的是雷达图的layout
2. r是**长度**，theta是**角度**。这个是*数学知识*，用的是**弧度制**
3. `set_thetagrids`用的度数体系和之前的`theta`是**不同**的。是**角度制**。


```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
r = [1.5, 1.5, 1.5, 1.5, 1.5]
theta = [0, 0.5 * np.pi, np.pi, 1.5 * np.pi, 2 * np.pi]

ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)
angles = [x * 180 / np.pi for x in theta[:4]]
print(angles)
ax.set_thetagrids(angles, ["A", "B", "C", "D"])
ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
